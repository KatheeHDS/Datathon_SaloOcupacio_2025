# Clustering

```python
df = pd.read_csv("./df_sigledate.csv", encoding="latin1")
```

```python
df.head()
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
```

```python
df.dtypes
```

```python
print(df.columns.tolist())
```

```python
# Definición de columnas cluster
columnas_cluster = [
    "Consumo Comercial",
    "Consumo Domestico",
    "Consumo Industrial"
]

X = df[columnas_cluster]

# Normalización de datos
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Determinar número de clusters
inertia = []
K_range = range(1, 10)
for k in K_range:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(X_scaled)
    inertia.append(kmeans.inertia_)

plt.plot(K_range, inertia, 'bo-')
plt.xlabel('Número de clusters')
plt.ylabel('Inertia (Suma de distancias al cuadrado)')
plt.title('Método del codo para elegir k')
plt.show()
```

Se ve claramente que la tendencia a suavizar la curva es a partir de k=3

```python
df["Districte"].unique()
```

```python
# 4️⃣ Aplicar K-Means con el número elegido de clusters
k = 3  # por ejemplo, según el gráfico
kmeans = KMeans(n_clusters=k, random_state=42)
kmeans.fit(X_scaled)

# 5️⃣ Asignar cluster a cada fila
df["Cluster"] = kmeans.labels_

# 6️⃣ Resumen por cluster
resumen = df.groupby("Cluster")[columnas_cluster].mean()
print(resumen)
```

```python
df[["Districte", "Cluster"]]
```

```python
sns.pairplot(df, vars=["Consumo Comercial", "Consumo Domestico", "Consumo Industrial"], hue="Cluster", palette="Set2")
```

```python
df["Consumo Domestico (mill)"] = df["Consumo Domestico"] / 1e6
df["Consumo Industrial (mill)"] = df["Consumo Industrial"] / 1e6
df["Consumo Comercial (mill)"] = df["Consumo Comercial"] / 1e6
```

```python
plt.figure(figsize=(24,18))

sns.scatterplot(
    data=df,
    x="Consumo Domestico (mill)",
    y="Consumo Industrial (mill)",
    hue="Cluster",
    palette="Set2",
    s=100
)

# Añadir etiquetas con los distritos
for i, row in df.iterrows():
    plt.text(
        row["Consumo Domestico (mill)"],
        row["Consumo Industrial (mill)"],
        str(row["Districte"]),
        fontsize=8,
        ha='right',
        va='bottom'
    )

plt.xlabel("Consumo Doméstico (millones L)")
plt.ylabel("Consumo Industrial (millones L)")
plt.title("Clusters de distritos según consumo")
plt.grid(True)
plt.show()
```

Debido a la cantidad de elementos es imposible hacer ninguna distinción con rotulos. Por esto decidimos asignar colores a los diferentes distritos y poder identificarlos sin rotulos. El grafico que más interesante nos parece es el de Domestico/Industrial ya que son los dos sectores que más consumen


```python
plt.figure(figsize=(30,24))

sns.scatterplot(
    data=df,
    x="Consumo Domestico (mill)",
    y="Consumo Industrial (mill)",
    hue="Districte",
    palette="tab10",
    s=80
)

plt.xlabel("Consumo Doméstico (millones L)")
plt.ylabel("Consumo Industrial (millones L)")
plt.title("Distritos según consumo")
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.) 
plt.grid(True)
plt.show()
```
